# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
# split data into X and y
X=df.iloc[:,1:]
y=df['TARGET']
# fit model no training data
Boosting = GradientBoostingClassifier(n_estimators=600,learning_rate=0.1,max_depth= 4,subsample= 1.0)
Boosting.fit(X, y)
```

# Feature Importance

```python
feature_importance = Boosting.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
pos=pos[246:]
fig = plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.barh(pos, feature_importance[sorted_idx[246:]], align='center')
plt.yticks(pos, np.array(df.columns)[sorted_idx[:5]])
plt.title('Feature Importance')
```
